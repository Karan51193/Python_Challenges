# Question 2: Fibonacci Number, by FAANG

- The Fibonacci numbers, commonly denoted F(n) form a sequence, called the Fibonacci sequence, such that each number is the sum of the two preceding ones, starting from 0 and 1.
That is,
F(0) = 0, F(1) = 1,
F(n) = F(n -1) + F(n-2), for n > 1.
- Given n, calculate F(n).

```python
%%time

#Brute force
def fibo(n):
  if n>1:
    counter=0
    series=[0]*(n+1)
    while counter<n+1:
      if counter==0:
        series[counter]=0
      elif counter==1:
        series[counter]=1
      else:
        series[counter]=series[counter-1]+series[counter-2]
      counter+=1
    # print(series)
    return(series[n])
  else:
    print("Ensure n>1")

print(fibo(10))
```

```python
%%time

#Using recursion
def fibo(n):
  if n<=1:
    return n
  return fibo(n-1)+fibo(n-2)

print(fibo(10))
```

#Question 3: Check If Two String Arrays are Equivalent, by Facebook

- Given two string arrays word1 and word2, return true if the two arrays represent the same string, and false otherwise.
- A string is represented by an array if the array elements concatenated in order forms the string.


word1 = [“a”, “cb”], word2 = [“ab”, “c”]
https://leetcode.com/problems/check-if-two-string-arrays-are-equivalent/

```python
#Brute force method
def string_check(list_1,list_2):

  str1=" "
  str2=" "
  for i in list_1:
    str1+=i
  for i in list_2:
    str2+=i
  
  if str1==str2:
    return "true"
  else:
    return "false"


word1=["a","bc"]
word2=["ab","c"]

string_check(word1,word2)
```

#Question 4: Matrix Diagonal Sum, by Adobe

- Given a square matrix mat, return the sum of the matrix diagonals.
- Only include the sum of all the elements on the primary diagonal and all the elements on the secondary diagonal that are not part of the primary diagonal.
- https://leetcode.com/problems/matrix-diagonal-sum/

```python
#Solution 1

def mat_sum(mat):
  rows=len(mat)
  cols=len(mat[0])


  primary_elements=[]
  primary_indices=[]
  i=0
  j=0
  while i<rows and j<cols:#Include primary diagonal
    primary_elements.append(mat[i][j])
    primary_indices.append((i,j))
    i+=1
    j+=1
  
  secondary_elements=[]
  secondary_indices=[]
  i=0
  j=cols-1
  while j>-1:#Include secondary diagonal
    secondary_elements.append(mat[i][j])
    secondary_indices.append((i,j))
    i+=1
    j-=1


  final_sum=0

  final_sum=sum(primary_elements)

  # for l in secondary_elements:
  #   if l not in primary_elements:
  #     final_sum+=l

  for t in secondary_indices:
    if t not in primary_indices:
      final_sum+=mat[t[0]][t[1]]
      # print(t)



  return final_sum


mat = [[1,2,3],
              [4,5,6],
              [7,8,9]]
mat_sum(mat)
```

#Question 7:Gas Station
Given two integer arrays A and B of size N. There are N gas stations along a circular route, where the amount of gas at station i is A[i].

You have a car with an unlimited gas tank and it costs B[i] of gas to travel from station i to its next station (i+1). You begin the journey with an empty tank at one of the gas stations.

Return the minimum starting gas station's index if you can travel around the circuit once, otherwise return -1.

You can only travel in one direction. i to i+1, i+2, ... n-1, 0, 1, 2.. Completing the circuit means starting at i and ending up at i again.

https://www.interviewbit.com/problems/gas-station/

```python
def canCompleteCircuit(A, B):
  for i in range(len(A)):
    # print("i",i)
    station = i
    fuel=0
    successful_indices=[]
    start_station=station
    for j in range(len(A)):
      # print("j",j)
      # print("start station",start_station)
      # print("station",station)
      
      if station<len(A):#To check if station count exceeds length of array
        fuel+=A[station]#Fuel accumulated
        # print("Fuel",fuel)
        if fuel<0:#If  fuel not enough
          break
        else:
          fuel-=B[station]#Fuel consumed
          station+=1
          if j==(len(A)-1):
            successful_indices.append(start_station)
          continue
      else:#To check if station count exceeds length of array
        fuel+=A[station-len(A)]#Fuel accumulated
        
        # print("Fuel",fuel)

        if fuel<0:#If  fuel not enough
          break
        else:
          fuel-=B[station-len(A)]#Fuel consumed
          station+=1
          if j==(len(A)-1):
            successful_indices.append(start_station)
          continue
      


  # print(successful_indices)


  return min(successful_indices)

            

```

```python
%%time
def minion_game(string):

  kevin_score=0
  stuart_score=0

  vowels=["A","E","I","O","U"]

  for str_indx in range(len(string)):
    score=len(string)-str_indx
    if string[str_indx] in vowels:#If Substring starts with a vowel
      kevin_score+=score
    else:#If Substring starts with consonant
      stuart_score+=score

  if stuart_score>kevin_score:
    print(f"Stuart {stuart_score}")

  elif kevin_score>stuart_score:
    print(f"Kevin {kevin_score}")

  else:
    print("Draw")
```
